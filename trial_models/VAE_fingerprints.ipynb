```python
import os
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "1"

import cv2
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
from PIL import Image
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
import imageio.v2 as imageio

# Set memory growth for GPU (prevents OOM errors)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Load data
data = glob(r'C:\Users\shrut\OneDrive\Desktop\data/Db*/*')
print(f"Total images found: {len(data)}")

def read_images_optimized(data, target_size=(64, 64)):  # Reduced size for faster training
    """Read and preprocess images efficiently"""
    images = []
    print("Loading images...")
    
    for i, path in enumerate(data):
        if i % 500 == 0:
            print(f"Processed {i}/{len(data)} images")
            
        try:
            # Read image
            img = imageio.imread(path)
            
            # Convert to grayscale if needed
            if len(img.shape) == 3:
                img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            
            # Resize image
            img = cv2.resize(img, target_size)
            
            # Normalize to [0, 1]
            img = img.astype('float32') / 255.0
            
            images.append(img)
            
        except Exception as e:
            print(f"Error processing image {path}: {e}")
            continue
    
    return np.array(images)

# Load and preprocess images
print("Loading images...")
images_arr = read_images_optimized(data, target_size=(64, 64))  # Smaller size for faster training
print(f"Loaded images shape: {images_arr.shape}")

# Reshape for CNN input
images_arr = images_arr.reshape(-1, 64, 64, 1)
print(f"Reshaped images: {images_arr.shape}")
print(f"Data range: [{np.min(images_arr):.3f}, {np.max(images_arr):.3f}]")

# Split data
train_X, valid_X, train_ground, valid_ground = train_test_split(
    images_arr, images_arr, 
    test_size=0.2, 
    random_state=13
)

print(f"Training set: {train_X.shape}")
print(f"Validation set: {valid_X.shape}")

def create_optimized_autoencoder(input_shape=(64, 64, 1)):
    """Create a more efficient autoencoder architecture"""
    
    input_img = Input(shape=input_shape)
    
    # Encoder
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)  # 64x64x32
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2), padding='same')(x)  # 32x32x32
    
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)  # 32x32x64
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2), padding='same')(x)  # 16x16x64
    
    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)  # 16x16x128
    x = BatchNormalization()(x)
    encoded = MaxPooling2D((2, 2), padding='same')(x)  # 8x8x128
    
    # Decoder
    x = Conv2D(128, (3, 3), activation='relu', padding='same')(encoded)  # 8x8x128
    x = BatchNormalization()(x)
    x = UpSampling2D((2, 2))(x)  # 16x16x128
    
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)  # 16x16x64
    x = BatchNormalization()(x)
    x = UpSampling2D((2, 2))(x)  # 32x32x64
    
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)  # 32x32x32
    x = BatchNormalization()(x)
    x = UpSampling2D((2, 2))(x)  # 64x64x32
    
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)  # 64x64x1
    
    return Model(input_img, decoded)

# Create model
autoencoder = create_optimized_autoencoder()
autoencoder.compile(
    optimizer=Adam(learning_rate=0.001),  # Adam is generally faster than RMSprop
    loss='mse',
    metrics=['mae']
)

print(autoencoder.summary())

# Create directory for checkpoints if it doesn't exist
checkpoint_dir = 'checkpoints'
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)

# Custom callback to save checkpoints every 40 epochs
class CustomCheckpoint(keras.callbacks.Callback):
    def __init__(self, save_freq=40, checkpoint_dir='checkpoints'):
        super().__init__()
        self.save_freq = save_freq
        self.checkpoint_dir = checkpoint_dir
    
    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.save_freq == 0:
            checkpoint_path = os.path.join(self.checkpoint_dir, f'autoencoder_epoch_{epoch+1:03d}.h5')
            self.model.save(checkpoint_path)
            print(f"\nCheckpoint saved at epoch {epoch+1}: {checkpoint_path}")

# Callbacks for training (removed EarlyStopping to ensure 200 epochs)
callbacks = [
    CustomCheckpoint(save_freq=40, checkpoint_dir=checkpoint_dir),
    ReduceLROnPlateau(factor=0.5, patience=10, min_lr=1e-7, verbose=1)
]

# Training parameters
batch_size = 32  # Reduced batch size for memory efficiency
epochs = 200     # Exactly 200 epochs as requested

print("Starting training for 200 epochs with checkpoints every 40 epochs...")
history = autoencoder.fit(
    train_X, train_ground,
    batch_size=batch_size,
    epochs=epochs,
    verbose=1,
    validation_data=(valid_X, valid_ground),
    callbacks=callbacks
)

# Plot training history
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Training MAE')
    plt.plot(history.history['val_mae'], label='Validation MAE')
    plt.title('Model MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

plot_training_history(history)

# Test the autoencoder
def visualize_reconstruction(autoencoder, test_images, n_images=5):
    """Visualize original vs reconstructed images"""
    
    # Get predictions
    reconstructed = autoencoder.predict(test_images[:n_images])
    
    plt.figure(figsize=(15, 6))
    for i in range(n_images):
        # Original
        plt.subplot(2, n_images, i + 1)
        plt.imshow(test_images[i].reshape(64, 64), cmap='gray')
        plt.title('Original')
        plt.axis('off')
        
        # Reconstructed
        plt.subplot(2, n_images, i + 1 + n_images)
        plt.imshow(reconstructed[i].reshape(64, 64), cmap='gray')
        plt.title('Reconstructed')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Visualize results
print("Visualizing reconstruction results...")
visualize_reconstruction(autoencoder, valid_X, n_images=5)

# Calculate reconstruction error
def calculate_reconstruction_metrics(autoencoder, test_images):
    """Calculate various reconstruction metrics"""
    reconstructed = autoencoder.predict(test_images)
    
    mse = np.mean((test_images - reconstructed) ** 2)
    mae = np.mean(np.abs(test_images - reconstructed))
    
    print(f"Reconstruction MSE: {mse:.6f}")
    print(f"Reconstruction MAE: {mae:.6f}")
    
    return mse, mae

calculate_reconstruction_metrics(autoencoder, valid_X)

# Save the final model
final_model_path = 'fingerprint_autoencoder_final_200epochs.h5'
autoencoder.save(final_model_path)
print(f"Final model saved: {final_model_path}")

# Print checkpoint summary
print("\nTraining completed!")
print("Checkpoints saved at epochs: 40, 80, 120, 160, 200")
print(f"All checkpoints are saved in the '{checkpoint_dir}' directory")
print(f"Final model saved as: {final_model_path}")

# Optional: If you want to scale up to larger images later
def create_larger_autoencoder(input_shape=(128, 128, 1)):
    """Create autoencoder for larger images (use only after confirming the smaller version works)"""
    
    input_img = Input(shape=input_shape)
    
    # Encoder
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2))(x)  # 64x64
    
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)  # 32x32
    
    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)  # 16x16
    
    x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    encoded = MaxPooling2D((2, 2))(x)  # 8x8
    
    # Decoder
    x = Conv2D(256, (3, 3), activation='relu', padding='same')(encoded)
    x = UpSampling2D((2, 2))(x)  # 16x16
    
    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = UpSampling2D((2, 2))(x)  # 32x32
    
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = UpSampling2D((2, 2))(x)  # 64x64
    
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = UpSampling2D((2, 2))(x)  # 128x128
    
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)
    
    return Model(input_img, decoded)
```
